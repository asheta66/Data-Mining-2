```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read data from CSV file to DataFrame
df = pd.read_excel("ClevlandDataSet.xlsx")
```

```python
# Display data
display(df.head())
```

```python
# DataFrame info
df.info()
```

```python
# Get column names
print("Columns:", df.columns.tolist())
```

```python
# Allow user to pick a column
col = input("Enter the column name to work with: ")
while col not in df.columns:
    print("Invalid column name. Please choose from:", df.columns.tolist())
    col = input("Enter the column name: ")
```

```python
# Check for null values
print(f"Null values in {col}: {df[col].isnull().sum()}")
```

```python
# Check for NA values
print(f"NA values in {col}: {df[col].isna().sum()}")
```

```python
# Get unique values
print(f"Unique values in {col}: {df[col].unique()}")
```

```python
# Use len
print(f"Length of {col}: {len(df[col])}")
```

```python
# Use count
print(f"Count of non-null entries in {col}: {df[col].count()}")
```

```python
# Plot bar chart
plt.figure(figsize=(5, 4))
df[col].value_counts().plot(kind='bar')
plt.xticks(rotation=45)
plt.xlabel(col)
plt.ylabel('Count')
plt.title(f'Bar Chart of {col}')
plt.grid(True)
plt.show()
```

```python
# Plot pie chart
plt.figure(figsize=(5, 4))
df[col].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title(f'Pie Chart of {col}')
plt.ylabel('')
plt.show()
```

```python
# Seaborn visualization
plt.figure(figsize=(5, 4))
sns.countplot(x=col, data=df)
plt.xticks(rotation=45)
plt.xlabel(col)
plt.ylabel('Count')
plt.title(f'Seaborn Count Plot of {col}')
plt.grid(True)
plt.show()
```

```python
# Seaborn visualization with different bar colors
plt.figure(figsize=(5, 4))
palette = sns.color_palette('hsv', len(df[col].unique()))
sns.countplot(x=col, data=df, hue=col, palette=palette, legend=False)
plt.xticks(rotation=45)
plt.xlabel(col)
plt.ylabel('Count')
plt.title(f'Seaborn Count Plot of {col}')
plt.grid(True)
plt.show()
```
