**Handling Missing Data**

Missing data is a common problem in real-world datasets and can negatively impact analysis and machine learning model performance. Handling missing data appropriately ensures data integrity and improves model accuracy.

**Deletion Method**

We remove rows or columns that contain missing values.

```python
import pandas as pd

# Creating a dataset with missing values
data2 = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, None, 40, 28],  # Missing Age for Charlie
    'Salary': [50000, 60000, 55000, None, 58000]  # Missing Salary for David
}

df2 = pd.DataFrame(data2)
print("Original Dataset:\n")
df2
```

```python
# **2. Deleting Columns with Missing Values**

df2_deleted_columns = df2.dropna(axis=1)  # Drops columns instead of rows
print("\nDataset after Dropping Columns with Missing Values:\n")
df2_deleted_columns
```

**Fill with Mean, Median, or Mode**

```python
import pandas as pd

# Creating a dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, None, 40, 28],  # Missing Age for Charlie
    'Salary': [50000, 60000, 55000, None, 58000],  # Missing Salary for David
    'City': ['New York', 'Los Angeles', None, 'Chicago', 'Houston']  # Missing City for Charlie
}

df = pd.DataFrame(data)
print("Original Dataset:\n")
df
```

```python
# **Calculating Means for Numeric Columns with Missing Values**
mean_age = df['Age'].mean()
mean_salary = df['Salary'].mean()

print("\nMean Age:", mean_age)
print("Mean Salary:", mean_salary)
```

```python
# Filling missing values with Mean (For Numeric Columns)
df_mean = df.copy()
df_mean = df_mean.assign(
    Age=df_mean['Age'].fillna(df_mean['Age'].mean()),
    Salary=df_mean['Salary'].fillna(df_mean['Salary'].mean()),
    City=df_mean['City'].fillna(df_mean['City'].mode()[0])  # Mode for categorical column
)

print("\nDataset after Mean Imputation:\n")
df_mean
```

**Median**

```python
import pandas as pd

# Creating a dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, None, 40, 28],  # Missing Age for Charlie
    'Salary': [50000, 60000, 55000, None, 58000],  # Missing Salary for David
    'City': ['New York', 'Los Angeles', None, 'Chicago', 'Houston']  # Missing City for Charlie
}

df = pd.DataFrame(data)
print("Original Dataset:\n")
df
```

```python
# **Calculating Medians for Numeric Columns with Missing Values**
median_age = df['Age'].median()
median_salary = df['Salary'].median()

print("\nMedian Age:", median_age)
print("Median Salary:", median_salary)
```

```python
# Creating a copy of the dataset for Median Imputation
df_median = df.copy()

# **Applying Median Imputation**
df_median['Age'] = df_median['Age'].fillna(df_median['Age'].median())
df_median['Salary'] = df_median['Salary'].fillna(df_median['Salary'].median())

# **Applying Mode Imputation for Categorical Column (City)**
df_median['City'] = df_median['City'].fillna(df_median['City'].mode()[0])

print("\nDataset after Median and Mode Imputation:\n")
df_median
```

**Data Normalization and Data Standardization**

**Normalization** scales the data into a specific range, typically [0, 1] or [-1, 1]. This is useful when the features have different units or magnitudes.

**Standardization** transforms the data to have a mean of 0 and a standard deviation of 1, which is useful when the features have different scales.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Generate sample data (10 data points, 2 features)
data = np.array([[10, 200],
                 [20, 250],
                 [30, 300],
                 [40, 350],
                 [50, 400],
                 [60, 450],
                 [70, 500],
                 [80, 550],
                 [90, 600],
                 [100, 650]])

# Convert to DataFrame for better visualization
df = pd.DataFrame(data, columns=['Feature1', 'Feature2'])

# Display original data
print("Original Data:")
df
```

```python
# --- Normalization (Min-Max Scaling) ---
scaler_minmax = MinMaxScaler()
normalized_data = scaler_minmax.fit_transform(df)

# Convert normalized data to DataFrame for better visualization
df_normalized = pd.DataFrame(normalized_data, columns=['Feature1', 'Feature2'])

# Display normalized data
print("\nNormalized Data (Min-Max Scaling):")
df_normalized
```

```python
# --- Standardization (Z-score Scaling) ---
scaler_standard = StandardScaler()
standardized_data = scaler_standard.fit_transform(df)

# Convert standardized data to DataFrame for better visualization
df_standardized = pd.DataFrame(standardized_data, columns=['Feature1', 'Feature2'])

# Display standardized data
print("\nStandardized Data (Z-score Scaling):")
df_standardized
```
