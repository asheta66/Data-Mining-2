```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# Sample dataset
data = {
    'Time': ['08:00', '09:00', '10:00', '11:00', '12:00', '13:00', '14:00', '15:00', '16:00', '17:00'],
    'Slowness in traffic (%)': ['30,5', '45,2', '25,3', '50,1', '40,0', '35,8', '20,4', '48,7', '55,3', '60,1']
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Clean data: Convert 'Slowness in traffic (%)' to float
df['Slowness in traffic (%)'] = df['Slowness in traffic (%)'].str.replace(',', '.').astype(float)

# Display the first few rows
print("Dataset Preview:")
print(df.head())

# Visualizing the data
plt.figure(figsize=(8, 4))
plt.bar(df['Time'], df['Slowness in traffic (%)'], color='skyblue')
plt.title('Slowness in Traffic (%) by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Slowness in Traffic (%)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# K-means clustering
X = df[['Slowness in traffic (%)']]
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# Visualizing clusters
plt.figure(figsize=(8, 4))
sns.scatterplot(x=df['Time'], y=df['Slowness in traffic (%)'], hue=df['Cluster'], palette='viridis', s=100)
plt.title('Traffic Clusters')
plt.xlabel('Time of Day')
plt.ylabel('Slowness in Traffic (%)')
plt.xticks(rotation=45)
plt.legend(title='Cluster')
plt.show()

# Summary of clusters
print("Clustered Data:")
print(df)
```
