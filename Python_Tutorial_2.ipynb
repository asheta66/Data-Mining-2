**1. NumPy — Arrays, Matrix Operations, Statistical Functions**

```python
import numpy as np

# Creating arrays
array = np.array([1, 2, 3, 4, 5])
matrix = np.array([[1, 2, 3], [4, 5, 6]])

print("Array:", array)
print("Matrix:\n", matrix)
```

```python
# Matrix operations
transpose = matrix.T
dot_product = np.dot(matrix, matrix.T)

print("Transpose:\n", transpose)
print("Dot Product:\n", dot_product)
```

```python
# Statistical functions
mean = np.mean(array)
std_dev = np.std(array)
sum_elements = np.sum(array)

print("Mean:", mean)
print("Standard Deviation:", std_dev)
print("Sum:", sum_elements)
```

**2. Pandas — DataFrames, Reading/Writing Data, Basic Manipulation**

```python
import pandas as pd

# Creating a DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Score": [85, 90, 78]
}

df = pd.DataFrame(data)
print("DataFrame:\n", df)
```

```python
# Reading and writing data
# df = pd.read_csv('data.csv')  # reading a CSV file
```

```python
# Basic manipulations
df['Passed'] = df['Score'] > 80
print("Updated DataFrame:\n", df)
```

```python
# Descriptive statistics
print("Summary Statistics:\n", df.describe())
```

```python
# Filtering rows
passed_students = df[df['Passed']]
print("Students who passed:\n", passed_students)
```

**3. Matplotlib — Scatter Plots, Histograms**

```python
import matplotlib.pyplot as plt

# Sample data
x = [10, 20, 30, 40, 50]
y = [15, 25, 35, 45, 55]

# Scatter plot
plt.figure(figsize=(4, 3))
plt.scatter(x, y, color='blue', marker='o')
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()
```

```python
# Histogram
scores = [85, 90, 78, 85, 88, 92, 75, 90, 82]
plt.figure(figsize=(4, 3))
plt.hist(scores, bins=5, color='green', edgecolor='black')
plt.title("Histogram of Scores")
plt.xlabel("Score Ranges")
plt.ylabel("Frequency")
plt.show()
```

**4. Seaborn — Advanced Visualizations (Heatmaps, Pair Plots)**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
tips = sns.load_dataset('tips')
print(tips.head())

# Pair plot — visualize relationships between numeric variables
plt.figure(figsize=(5, 5))
sns.pairplot(tips, hue='sex', palette='coolwarm')
plt.show()
```

```python
# Heatmap — correlation matrix
plt.figure(figsize=(4, 3))
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='Blues', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
