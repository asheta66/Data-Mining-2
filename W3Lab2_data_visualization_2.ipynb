```python
import matplotlib.pyplot as plt

month_number = [1, 2, 3, 4, 5, 6, 7]
new_cases = [9926, 76246, 681488, 2336640, 2835147, 4226655, 6942042]

fig = plt.figure(figsize=(6,3))
plt.plot(month_number, new_cases)
plt.title('New Reported Cases By Month (Globally)')
plt.xlabel('Month Number')
plt.ylabel('Number Of Cases')
plt.ticklabel_format(axis='y', style='plain')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Capital Bikeshare dataset
bike_sharing = pd.read_csv('day.csv')
```

```python
bike_sharing.head()
```

```python
fig = plt.figure(figsize=(6,4))
# Create a scatter plot to explore the relationship between temperature and bikes rented
plt.scatter(bike_sharing['temp'], bike_sharing['cnt'])
plt.xlabel('Temperature (Normalized)')
plt.ylabel('Bikes Rented')
plt.grid(True)
plt.show()
```

```python
# Group the data by weekday, calculate the mean, and select 'casual' and 'registered' rental columns
weekday_averages = bike_sharing.groupby('weekday').mean(numeric_only=True)[['casual', 'registered']].reset_index()

fig = plt.figure(figsize=(6,3))
# Create a bar plot for the average number of registered rentals per weekday
plt.bar(weekday_averages['weekday'], weekday_averages['registered'])

# Customize the x-axis with day names and rotate labels for better readability
plt.xticks(
    ticks=[0, 1, 2, 3, 4, 5, 6],
    labels=['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'],
    rotation=30)
plt.show()
```

```python
import pandas as pd

# Create a sample dataset
data = {
    'Time': ['08:00', '09:00', '10:00', '11:00', '12:00', '13:00', '14:00', '15:00', '16:00', '17:00'],
    'Slowness in traffic (%)': ['30,5', '45,2', '25,3', '50,1', '40,0', '35,8', '20,4', '48,7', '55,3', '60,1']
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Clean 'Slowness in traffic (%)' column and convert to float
df['Slowness in traffic (%)'] = df['Slowness in traffic (%)'].str.replace(',', '.').astype(float)

# Plot the bar chart
plt.figure(figsize=(10, 4))
plt.bar(df['Time'], df['Slowness in traffic (%)'], color='skyblue')
plt.title('Slowness in Traffic (%) by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Slowness in Traffic (%)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
