```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Dataset
data = {
    'Age': ['Young', 'Young', 'Middle-aged', 'Middle-aged', 'Old', 'Old'],
    'Income': ['Low', 'High', 'Low', 'High', 'Low', 'High'],
    'Buy': ['No', 'Yes', 'No', 'Yes', 'Yes', 'Yes']
}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert categorical data to numerical values
df['Age'] = df['Age'].map({'Young': 0, 'Middle-aged': 1, 'Old': 2})
df['Income'] = df['Income'].map({'Low': 0, 'High': 1})
df['Buy'] = df['Buy'].map({'No': 0, 'Yes': 1})

# Features and target variable
X = df[['Age', 'Income']]
y = df['Buy']

# Create and train the Decision Tree Classifier
clf = DecisionTreeClassifier(criterion='gini')
clf.fit(X, y)

# Plot the tree
plt.figure(figsize=(10, 8))
plot_tree(clf,
          feature_names=['Age', 'Income'],
          class_names=['No', 'Yes'],
          filled=True,
          rounded=True,
          fontsize=12)
plt.show()
```
