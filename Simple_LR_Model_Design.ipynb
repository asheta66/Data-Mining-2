```python
import numpy as np
import matplotlib.pyplot as plt

# Given new data points with a larger spread
x = np.array([1, 3, 7, 10, 13, 16, 9, 3, 12, 15, 22, 26])
y = np.array([5, 7, 5, 12, 23, 29, 35, 32, 48, 22, 33, 23])

# Calculate the mean of x and y
x_bar = np.mean(x)
y_bar = np.mean(y)

# Calculate the sums S_xx, S_xy, and S_yy
S_xx = np.sum((x - x_bar) ** 2)
S_xy = np.sum((x - x_bar) * (y - y_bar))
S_yy = np.sum((y - y_bar) ** 2)

# Calculate the least squares estimates for beta_1 and beta_0
beta_1 = S_xy / S_xx
beta_0 = y_bar - beta_1 * x_bar

# Print the calculated values with 3 decimal places
print(f"beta_1 = {beta_1:5.3f}")
print(f"beta_0 = {beta_0:5.3f}")

# Calculate the fitted line
y_fitted = beta_0 + beta_1 * x

# Plot the data points and the fitted line
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, y_fitted, color='red', label=f'Fitted line: y = {beta_0:5.3f} + {beta_1:5.3f}x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fitted Line')
plt.legend()
plt.grid(True)
plt.show()
```
