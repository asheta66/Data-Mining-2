```python
import numpy as np
import matplotlib.pyplot as plt

# Given data points
x = np.array([0, 1, 6, 3, 5])
y = np.array([6, 5, 2, 4, 3])

# Step 1: Calculate the mean of x and y
x_bar = np.mean(x)
y_bar = np.mean(y)

# Step 2: Calculate Sxx, Sxy, and Syy
Sxx = np.sum((x - x_bar)**2)
Sxy = np.sum((x - x_bar) * (y - y_bar))
Syy = np.sum((y - y_bar)**2)

# Step 3: Calculate the least squares estimates for beta_1 and beta_0
beta_1 = Sxy / Sxx
beta_0 = y_bar - beta_1 * x_bar

# Step 4: Plot the scatter diagram and the fitted line
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, beta_0 + beta_1 * x, color='red', label=f'Fitted line: y = {beta_0:.2f} + {beta_1:.2f}x')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression: Best Fit Line')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

# Display the calculated values
print(f"Mean of x: {x_bar}")
print(f"Mean of y: {y_bar}")
print(f"Sxx: {Sxx}")
print(f"Sxy: {Sxy}")
print(f"Syy: {Syy}")
print(f"Beta_1 (Slope): {beta_1}")
print(f"Beta_0 (Intercept): {beta_0}")
```
