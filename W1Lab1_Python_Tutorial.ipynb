```python
## Python Basics: Colab Notebook

# --- Basic Syntax and Data Types ---

## Variables and Data Types
# Integer
a = 10
print("Integer:", a, type(a))

# Float
b = 10.5
print("Float:", b, type(b))

# String
c = "Hello, World!"
print("String:", c, type(c))
```

```python
# Boolean
d = True
print("Boolean:", d, type(d))

## Type Conversion and Casting
x = int("5")
y = float("10.5")
z = str(100)
print(x, type(x))
print(y, type(y))
print(z, type(z))
```

```python
# --- Operators and Expressions ---

## Arithmetic Operators
print("Addition:", 5 + 3)
print("Subtraction:", 5 - 3)
print("Multiplication:", 5 * 3)
print("Division:", 5 / 3)
print("Floor Division:", 5 // 3)
print("Modulus:", 5 % 3)
print("Exponentiation:", 5 ** 3)
```

```python
## Comparison Operators
print("Equal:", 5 == 3)
print("Not Equal:", 5 != 3)
print("Greater than:", 5 > 3)
print("Less than:", 5 < 3)
```

```python
## Logical Operators
print("AND:", True and False)
print("OR:", True or False)
print("NOT:", not True)
```

```python
## Assignment Operators
x = 5
x += 3
print("+= Operator:", x)
x -= 2
print("-= Operator:", x)

## Operator Precedence
result = 5 + 2 * 3
print("Operator Precedence Result:", result)
```

```python
# --- Control Flow ---

## If Statements
num = 10
if num > 0:
    print("Positive number")
```

```python
## If-Elif-Else
num = 0
if num > 0:
    print("Positive")
elif num == 0:
    print("Zero")
else:
    print("Negative")
```

```python
## Nested Conditions
num = 15
if num > 10:
    if num < 20:
        print("Number is between 10 and 20")
```

```python
## The pass Statement
x = 100
if x > 50:
    pass  # Placeholder for future code
```

```python
## For Loops
print("For Loop Example:")
for i in range(5):
    print(i)
```

```python
## While Loops
print("While Loop Example:")
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
## Break, Continue, Else in Loops
print("Break Example:")
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
print("Continue Example:")
for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
print("Else in Loops:")
for i in range(5):
    print(i)
else:
    print("Loop finished")
```

```python
## Looping through Lists and Strings
print("Looping through a list:")
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
print("Looping through a string:")
for char in "Python":
    print(char)
```

```python
def example_function():
    print("Hello, this is my first function.")

example_function()
```

```python
## Function Arguments
def add(a, b):
    return a + b

print("Sum:", add(5, 3))
```

```python
## Positional, Keyword, and Default Arguments
def describe_pet(pet_name, animal_type="dog"):
    print(f"I have a {animal_type} named {pet_name}.")

describe_pet("Buddy")
describe_pet(pet_name="Kitty", animal_type="cat")
```

```python
## Return Values
def square(num):
    return num * num

print("Square:", square(4))
```

```python
## Multiple Return Values
def get_coordinates():
    return 10, 20

x, y = get_coordinates()
print("Coordinates:", x, y)
```

```python
## Lambda Functions
square = lambda x: x ** 2
print("Lambda Square:", square(5))
```
